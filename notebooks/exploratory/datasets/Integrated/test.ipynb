```python
import pandas as pd
```

```python
from src.features import get_raw_microbe_diseases, get_diseases, get_microbes

raw_associations = pd.DataFrame(get_raw_microbe_diseases())
disease_data = pd.DataFrame(get_diseases())
microbe_data = pd.DataFrame(get_microbes())
```

```python
s = 0
for i in range(len(raw_associations)):
    b = False
    for j in range(len(microbe_data)):
        if b:
            break
        for k in range(1, 8):
            if raw_associations.iloc[i, 1] in microbe_data.iloc[j, k]:
                b = True
                s += 1
                print('yes')
                break
    if not b: 
        print(raw_associations.iloc[i, 1], 'no')
        
print(f'we had {s} yes from {len(raw_associations)}')
```

```python
s = 0

for i in range(len(raw_associations)):
    b = False
    for j in range(len(disease_data)):
        if b:
            break
        if raw_associations.iloc[i, 0] in disease_data.iloc[j, 2]:
            b = True
            s += 1
            print('yes')
    if not b: 
        print(raw_associations.iloc[i, 0], 'no')
        
print(f'we had {s} yes from {len(raw_associations)}')
```
